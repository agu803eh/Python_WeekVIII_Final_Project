```python
countries_of_interest = ['United States', 'Brazil']
filtered_data = data[data['location'].isin(countries_of_interest)]
print(filtered_data)
```

```python
filtered_data = filtered_data[['date','iso_code', 'location', 'total_cases', 'total_deaths', 'total_vaccinations','new_cases','new_deaths']]
print(filtered_data)
```

```python
filtered_data = filtered_data.dropna(subset=['date', 'location', 'total_cases', 'total_deaths', 'total_vaccinations','new_cases','new_deaths','iso_code'])
print (filtered_data)
```

```python
filtered_data['total_cases'] = filtered_data['total_cases'].fillna(method='ffill')
filtered_data['total_vaccinations'] = filtered_data['total_vaccinations'].fillna(method='ffill')
filtered_data['new_cases'] = filtered_data['new_cases'].fillna(method='ffill')
filtered_data['new_deaths'] = filtered_data['new_deaths'].fillna(method='ffill')
filtered_data['total_deaths'] = filtered_data['total_deaths'].fillna(method='ffill')
filtered_data['iso_code'] = filtered_data['iso_code'].fillna(method='ffill')
print (filtered_data)
```

```python
# Save the filtered data to a new CSV file
filtered_data.to_csv('filtered_covid_data2.csv', index=False)
```

```python
# Print the first few rows of the filtered data
print(filtered_data.head())
```

```python
filtered_data['date'] = pd.to_datetime(filtered_data['date'])
print(filtered_data)
# Sort the data by 'date'
filtered_data = filtered_data.sort_values(by='date')
```

```python
# Print data specifically for 'Brazil'
brazil_data = filtered_data[filtered_data['location'] == 'Brazil']
print(brazil_data)
```

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot total cases over time for Brazil and the United States
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='date', y='total_cases', hue='location')
plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('total_cases_over_time.png')
plt.show()
```

```python
plt.figure(figsize=(14, 7))

# Plot each country separately
for country in filtered_data['location'].unique():
    country_data = filtered_data[filtered_data['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('total_deaths_over_time.png')
plt.show()
```

```python
# Compare daily new cases for Brazil and the United States
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='date', y='new_cases', hue='location')
plt.title('Daily New COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('new_cases_over_time.png')
plt.show()
```

```python
# Calculate the death rate: total_deaths/total_cases
filtered_data['death_rate'] = filtered_data['total_deaths'] / filtered_data['total_cases']
# Plot the death rate over time for Brazil and the United States
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='date', y='death_rate', hue='location')
plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('death_rate_over_time.png')
plt.show()
```

```python
# Plot cumulative vaccinations over time for Brazil and the United States
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='date', y='total_vaccinations', hue='location')
plt.title('Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('vaccinations_over_time.png')
plt.show()
```

```python
# Compare % vaccinated population against % of unvaccinated population using a pie chart
# calculate the percentage of vaccinated population by dividing total_vaccinations by total_cases
filtered_data['vaccination_percentage'] = (filtered_data['total_vaccinations'] / filtered_data['total_cases']) * 100
# calculate the percentage of unvaccinated population
filtered_data['unvaccinated_percentage'] = 100 - filtered_data['vaccination_percentage']
# Compare % vaccinated population against % of unvaccinated population using a pie chart for brazil and the united states
brazil_data = filtered_data[filtered_data['location'] == 'Brazil']
us_data = filtered_data[filtered_data['location'] == 'United States']
print(brazil_data)
print(us_data)
```

```python
#Compare %vaccinated population of brazil and the united states
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='date', y='vaccination_percentage', hue='location')
plt.title('COVID-19 Vaccination Percentage Over Time')
plt.xlabel('Date')
plt.ylabel('Vaccination Percentage')
plt.xticks(rotation=45)
plt.legend(title='Country')
plt.tight_layout()
plt.savefig('vaccination_percentage_over_time.png')
plt.show()
```
